```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_excel('Datafile/data.xlsx')              
```

```python
df.head()
```

## Line Chart

```python
df['Date'] =pd.to_datetime(df['Date'])# convert to datetime format
#plotting the data
plt.plot(df['Date'], df['Sales'], color='violet')#Changing the graph to another color
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)#rotating the x axis to 45 degrees
plt.show()
```

## Barchat

```python
plt.bar(df['Product'], df['Sales'], color='#827272')#Changing the graph to another color
plt.title('Product sales')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

```python
cars = pd.read_excel('Datafile/Car_data.xlsx')
```

```python
cars.head()
```

## Scatter Plot
Knowing the relationship between two variables(correlation)
### Positive correlation
one variable increases making the related variable increasing
### Negative correlation
one variable decreases making the related variable decreasing

```python
plt.figure(figsize=(10,6))
plt.scatter(cars['Max Price (Lakh)'],cars['Insurance'], color='green')
       #calculate regression line
slope, intercept=np.polyfit(cars['Max Price (Lakh)'], cars['Insurance'], 1)
       #generate x values for the regression line
x_line =np.linspace(cars['Max Price (Lakh)'].min(), cars['Max Price (Lakh)'].max(),100)
       #generate y values for the regression line
y_line = slope*x_line + intercept
       #plot the regression line   
plt.plot(x_line, y_line, color='black', label ='regression line')
plt.title('Max Price (Lakh) Over Insurance')
plt.xlabel('Max Price (Lakh)')
plt.ylabel('Insurance')
plt.grid(True)
plt.show()
```

## Pie Chart

```python
#aggregate data by car model
sales_data = df.groupby('Category')['Sales'].sum()
sales_data
```

```python
#Create the pie chart
plt.figure(figsize =(8, 8))
plt.pie(sales_data, labels=sales_data.index, autopct='%1.1f%%')
plt.title('Proportion of Sales by Category')
plt.show()
```

```python
corn = df.groupby('Country')['Sales'].sum()
corn
```

```python
plt.figure(figsize =(8, 8))
plt.pie(corn, labels=corn.index, autopct='%1.1f%%')
plt.title('Proportion of Sales by Country')
plt.show()
```

## Histogram

```python
plt.hist(df['Sales'], bins=5, color='black', edgecolor='pink')
plt.title('Sales Frequency')
plt.xlabel('bins')
plt.ylabel('frequencies')
plt.show()
```

```python

plt.fill_between(df['Date'], df['Sales'], color='lightgreen', alpha = 1)#alpha changes the saturation of colour
plt.plot(df['Date'], df['Sales'], color='green')
plt.title('Cumulative Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)#rotating the x axis to 45 degrees
plt.show()
```
