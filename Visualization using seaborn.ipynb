```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
car = pd.read_excel('Datafile/Car_data.xlsx')
car.head()
```

```python
variables = ['Min Price (Lakh)',	'Max Price (Lakh)',	'Range (kmpl)',	'CC',	'Seats',	'Variants',	'Type',	'Ex-Showroom Price',	'RTO',
           'Insurance',	'Other',	'Onroad Price']
```

```python
car.columns
```

```python
plt.figure(figsize=(15,12))
for i, var in enumerate(variables, start=1):
    plt.subplot(4, 3, i)#specifying how to place the subplots
    sns.scatterplot(x=car[var], y=car['Max Price (Lakh)'], alpha=0.7)
    plt.title(f'Scatter plot of {var} vs Onroad Price')
    plt.xlabel(var)
    plt.ylabel('Target')
plt.tight_layout()
plt.show()
```

```python
#Histogram for all independent variables
plt.figure(figsize=(15,10))
for i, var in enumerate(variables, start=1):
    plt.subplot(3, 4, i)#specifying how to place the subplots
    sns.histplot(x=car[var], bins=20, kde=True, color='orange')
    plt.title(f'Histogram of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.xticks(rotation=90)
plt.show()
```
